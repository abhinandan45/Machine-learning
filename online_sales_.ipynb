```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\abhinandan\Desktop\Machine_learning_full\Online Sales Data.csv')
```

```python
data.isnull().sum()
```

```python
data.drop(('Transaction ID'),axis=1,inplace=True)
```

```python
data
```

```python
f = data[data["Product Category"]=="Electronics"].groupby("Product Name")["Total Revenue"].sum().reset_index()
```

```python
b = data.groupby("Region")["Total Revenue"].sum().reset_index()
```

```python
c = b.index
```

```python
d = b.values
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=b, x="Region", y="Total Revenue", color="red")
plt.title("Total Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

```python
c = data.groupby("Payment Method")["Total Revenue"].sum().reset_index()

plt.title("Total revenue by payment method")
sns.barplot(data=c,x="Payment Method",y="Total Revenue",color="blue")
plt.xlabel("Payment Method")
plt.ylabel("Revenue")
plt.show()
```

```python
d = data.groupby("Product Category")["Units Sold"].sum().reset_index()
```

```python
plt.figure(figsize=(10 , 6))
plt.title("Product category by Unit sold")
sns.barplot(data=d,x="Product Category",y="Units Sold",color="green")
plt.xlabel("Product Category")
plt.ylabel("Unit Sold")
plt.show()
```

```python
e = data.groupby("Region")["Units Sold"].sum().reset_index()

plt.figure(figsize=(10 , 6))
plt.title("Unit sold in region")
sns.barplot(data=e,x="Region",y="Units Sold",color="orange")
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.show()
```

```python
f = data[data["Product Category"]=="Electronics"].groupby("Product Name")["Total Revenue"].sum().reset_index()
plt.figure(figsize=(10,7))
sns.barplot(data=f,x="Product Name",y="Total Revenue",color="green")
plt.show()
```
